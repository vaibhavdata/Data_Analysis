# how many times teams scores more than 200

```python
high_score=df.groupby(['match_id','inning','batting_team','bowling_team'])['total_runs'].sum().reset_index()
```

```python
high_score.head()
```

```python
more_200 =high_score[high_score['total_runs']>=200]
more_200.head()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(more_200['batting_team']);
```

```python
sns.countplot(more_200['bowling_team']);
```

 how many times team scores >200 runs against batting team

```python
more_200.set_index(['match_id'],inplace=True)
```

```python
high=more_200.rename(columns={"total_runs":"Count"})
high=high[high['Count']>200].groupby(['inning','batting_team','bowling_team']).count()
```

# Batsman Comparison

```python
df.columns
```

```python
balls=df.groupby(['batsman'])[['ball']].count().reset_index()
```

```python
balls
```

```python
runs=df.groupby(['batsman'])['batsman_runs'].sum().reset_index()
runs.columns=['bastsman','runs']
runs
```

# fours

```python
four=df.groupby(['batsman'])['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index()
four.columns=['bastman','4s']
```

```python
four
```

```python
six=df.groupby(['batsman'])['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index()
six.columns=["bastman","6s"]
six
```

```python
players=pd.concat([runs,balls.iloc[:,1],four.iloc[:,1],six.iloc[:,1]],axis=1)
players
```

```python
players['stick_rate']=(players['runs'] /players['ball'])*100
```

```python
players
```

```python
players.isna().sum()
```

```python
df.columns
```

```python
grp=df.groupby(['match_id','batsman','batting_team'])['batsman_runs'].sum().reset_index()
grp
```

```python
max=grp.groupby(['batsman'])['batsman_runs'].max().reset_index()
max.columns=['batsman','max_run']
max
```

```python
player1=pd.concat([players,max.iloc[:,1:3]],axis=1)
```
